```python
import numpy as np
from matplotlib import pyplot as plt
import matplotlib.image as mpimg
import pandas as pd
```

# Transformer-VAE

```python
ori_file = "level_0_Baseline_with0.npz"
ori_data = np.load(ori_file)

print(ori_data.files)
```

```python
gen_file = "/Users/zhaozheyuan/Desktop/mcs_project/timeVAE-main/outputs/gen_data/level_0_Baseline_with0/vae_int_level_0_Baseline_with0_prior_samples.npz"
gen_data = np.load(gen_file)

print(gen_data.files)
```

```python
abm_data = ori_data['data']

vae_data = gen_data['data']

print(abm_data.shape)
print(vae_data.shape)
```

```python
fig, axes = plt.subplots(2, 1, figsize=(6, 6))


for i in range(len(abm_data)):
    axes[0].plot(abm_data[i], color='lightblue', alpha=0.5)
axes[0].set_title('Level 0 Baseline Cum Infections')
axes[0].set_ylabel('Number of infections')


for i in range(len(vae_data)):
    axes[1].plot(vae_data[i], color='lightblue', alpha=0.5)
axes[1].set_title('VAE_INT Generated Cum Infections')
axes[1].set_ylabel('Number of infections')

plt.tight_layout()

plt.show()
```
